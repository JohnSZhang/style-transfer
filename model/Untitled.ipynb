```python
from gensim.models import Word2Vec
```

```python
sentences = []
```

```python
lines = [line.rstrip('\n') for line in open('./train.original.nltktok', 'r')]
org_sentences = [l.split() for l in lines]
```

```python
lines = [line.rstrip('\n') for line in open('./train.modern.nltktok', 'r')]
mod_sentences = [l.split() for l in lines]
```

```python
total_sentences = org_sentences + mod_sentences
```

```python
model = Word2Vec(sentences, size=192, min_count=1)
# summarize the loaded model
print(model)
# summarize vocabulary
words = list(model.wv.vocab)
print(words)
```

```python
import pickle
import numpy as np
```

```python
words = list(model.wv.vocab)
for w in words:
    embed[w] = model[w]
    
    
```
